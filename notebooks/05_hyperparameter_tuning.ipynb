```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Dense, LSTM, Input
```

```python
df = pd.read_csv('../data/TSLA.csv')

df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

data = df[['Close']]

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)

train_size = int(len(scaled_data) * 0.8)

train_data = scaled_data[:train_size]
test_data = scaled_data[train_size:]
```

```python
def create_sequences(data, seq_length=60):
    X, y = [], []
    
    for i in range(seq_length, len(data)):
        X.append(data[i-seq_length:i])
        y.append(data[i])
        
    return np.array(X), np.array(y)

X_train, y_train = create_sequences(train_data)
```

```python
def build_model(units=100):
    model = Sequential()
    
    model.add(Input(shape=(X_train.shape[1], 1)))
    
    model.add(LSTM(units, return_sequences=True))
    model.add(Dropout(0.2))
    
    model.add(LSTM(units))
    model.add(Dropout(0.2))
    
    model.add(Dense(1))
    
    model.compile(optimizer='adam', loss='mean_squared_error')
    
    return model
```

```python
model1 = build_model(units=50)

history1 = model1.fit(X_train, y_train, epochs=5, batch_size=32)

loss1 = history1.history['loss'][-1]
print("Experiment 1 Loss:", loss1)
```

```python
model2 = build_model(units=50)

history2 = model2.fit(X_train, y_train, epochs=10, batch_size=32)

loss2 = history2.history['loss'][-1]
print("Experiment 2 Loss:", loss2)
```

```python
model3 = build_model(units=100)

history3 = model3.fit(X_train, y_train, epochs=10, batch_size=32)

loss3 = history3.history['loss'][-1]
print("Experiment 3 Loss:", loss3)
```

```python
best_model = build_model(units=100)

best_model.fit(X_train, y_train, epochs=35, batch_size=32)

best_model.save('../models/lstm_tuned_best.keras')
```

```python
print("Final Comparison:")
print(f"Exp1 (5 epochs, 50 units): {loss1}")
print(f"Exp2 (10 epochs, 50 units): {loss2}")
print(f"Exp3 (10 epochs, 100 units): {loss3}")
```
